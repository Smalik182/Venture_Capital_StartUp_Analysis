```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

%matplotlib inline

vc_data = pd.read_csv('Dataset.csv')

# Since the dataset is long, we want to be able to see all column names
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

vc_data.head()
```

# Data Cleaning

Lets get rid of unnecessary columns first. 
We do not need URL's for the startups so we can first get rid of those.
We can also remove all the column with data that is not in USD, since there are columns with the conversion to USD.


```python

vc_data = vc_data.drop(['Organization Name URL', 'Acquired by URL','Twitter','Facebook','LinkedIn',
                       'Total Funding Amount', 'Total Funding Amount Currency',
                       'Acquisition Price', 'Price Currency', 'Unnamed: 20',
                       'Last Funding Amount Currency'], axis = 1)
vc_data.head(2)
```

### Date Columns

We can see that some of the column contain dates but are of object data type. Lets adjust those to DateTime Objects.

```python
vc_data["Last Funding Date"] = pd.to_datetime(vc_data['Last Funding Date'], format='%m/%d/%Y')
vc_data["Founded Date"] = pd.to_datetime(vc_data['Founded Date'], format='%m/%d/%Y')
vc_data["Announced Date"] = pd.to_datetime(vc_data['Announced Date'], format='%m/%d/%Y')
vc_data.info()
```

### Number of Employee column

We can see an error in the dataset where in some rows the Number of Employees is 'Nov-50' or '10-Jan'.
Lets also clean up the '$' and any newlines we may have (\n)

```python
clean_data = vc_data.replace({'Nov-50': np.nan, '10-Jan': np.nan, '\$' : '', '\n' : ''}) 

clean_data.head()
```

# Insights

Now lets begin on what insights we can get from this dataset. 

### Generally, how many founders does a start up have?

We can get a clearer picture of how many founders a startup has.
But first we need to get rid of rows where the number of founders is not available or is 0.

```python
# First we can replace all values that were empty win NaN using np.nan
clean_data['Number of Founders'].replace('', np.nan, inplace=True)
# Then we drop all values in the specified column using .dropna()
clean_num_founders = cleaned_data.dropna(subset=['Number of Founders'])

# Now we can address where the Number of Founders = 0
final_num_founders = clean_num_founders[clean_num_founders['Number of Founders'] != 0]

final_num_founders.head(5)
```

```python

num_of_founders = final_num_founders['Number of Founders'].value_counts()
plt.figure(figsize=(7, 6))
ax = num_of_founders.plot(kind='bar', rot=0, color="b")
ax.set_title("Bar Graph of Number of Founders", y = 1)
ax.set_xlabel('Number of Founders')
ax.set_ylabel('Frequency')
#ax.set_xticklabels(('Male', 'Female'))

for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    space = 1
    label = "{:.0f}".format(y_value)
    ax.annotate(label, (x_value, y_value), xytext=(0, space), textcoords="offset points", ha='center', va='bottom')
plt.show()
```
