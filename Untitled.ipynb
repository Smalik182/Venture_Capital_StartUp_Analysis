```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

%matplotlib inline

vc_data = pd.read_csv('Dataset.csv')

# Since the dataset is long, we want to be able to see all column names
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

vc_data.head()
#gc.enable()
```

# Data Cleaning

Lets get rid of unnecessary columns first. 
We do not need URL's for the startups so we can first get rid of those.
We can also remove all the column with data that is not in USD, since there are columns with the conversion to USD.


```python

vc_data = vc_data.drop(['Organization Name URL', 'Acquired by URL','Twitter','Facebook','LinkedIn',
                       'Total Funding Amount', 'Total Funding Amount Currency',
                       'Acquisition Price', 'Price Currency', 'Unnamed: 20',
                       'Last Funding Amount Currency'], axis = 1)# Note regex = True necessary
clean_data = vc_data.replace({'Nov-50': np.nan, '10-Jan': np.nan, '\$' : '', '\n' : ''}, regex = True) 

clean_data.head()
```

### Filtering by Industries

We can create a funtion that will alow the user to input multiple industries, the function will then run through the dataset and return all rows containing the users chosen industries. 

```python

## This function gets the user's industries
def user_Industries():
    num_of_industries = int(input('How many industries would you like to filter by?\n'))
    list_industries = []
   
    i = 0
    #while i < num_of_industries:
    for i in range(num_of_industries):
        industry = input("Enter the industry:\n")
        i += 1
        
        list_industries.append(industry)

    return list_industries

## We call the fuction here to store the input
a = user_Industries()



## We convert the list into a string
pattern = ', '.join(a)


## Here we are filtering the pattern through the rows of the column
def filter_Industries():
    mask = clean_data['Industries'].str.contains(pattern , case=False, na=False)
    final = clean_data[mask]
    return final

## Function Call
b =  filter_Industries()   

b.head(5)
```

# Insights

Now lets begin on what insights we can get from this dataset. 

### Generally, how many founders does a start up have?

We can get a clearer picture of how many founders a startup has.
But first we need to get rid of rows where the number of founders is not available or is 0.

```python
# Removing all rows where Number of Founders in missing
final_num_founders = clean_data[clean_data['Number of Founders'].notnull()].copy()

# Percentage of Number of Founders at start ups
clean_data['Number of Founders'].value_counts(normalize = True) * 100
```

We can see that majority of start ups have 1 (38.3%) or 2 (35.5%) founders. We can get a better idea of the number of founders by creating a frequency diagram:

```python
plt.style.use('fivethirtyeight')
num_of_founders = final_num_founders['Number of Founders'].value_counts()
plt.figure(figsize=(7, 6))
ax = num_of_founders.plot(kind='bar', rot=0, color="b")
ax.set_title("Bar Graph of Number of Founders", y = 1)
ax.set_xlabel('Number of Founders')
ax.set_ylabel('Frequency')
#ax.set_xticklabels(('Male', 'Female'))

for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    space = 1
    label = "{:.0f}".format(y_value)
    ax.annotate(label, (x_value, y_value), xytext=(0, space), textcoords="offset points", ha='center', va='bottom')
plt.show()
```
